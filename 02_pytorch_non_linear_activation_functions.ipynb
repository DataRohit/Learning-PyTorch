# **Replicating Non-Linear Activation Functions**

## **Imports**

```python
# Imports
import torch

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

## **Tensor**

```python
# Create a tensor
data_tensor = torch.arange(-10, 10, 1, dtype = torch.float32)
data_tensor
```

```python
# Visualize tensor
sns.lineplot(data_tensor)
plt.show()
```

## **ReLU**

```python
# Create a ReLU function manually
def relu(x):
    return torch.maximum(torch.tensor(0), x) # Input must be tensors

# Pass the tensor to ReLU function
relu(data_tensor)
```

```python
# Visualize ReLU activated tensor
sns.lineplot(relu(data_tensor))
plt.show()
```

## **Sigmoid**

```python
# Create sigmoid function manually
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

# Pass the tensor to the sigmoid function
sigmoid(data_tensor)
```

```python
# Visualize Sigmoid activated tensor
sns.lineplot(sigmoid(data_tensor))
plt.show()
```
